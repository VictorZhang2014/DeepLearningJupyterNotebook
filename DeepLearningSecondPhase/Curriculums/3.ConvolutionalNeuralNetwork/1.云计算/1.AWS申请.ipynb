# 2.AWS

<img src="../../../sources/img/CNN/amazonwebservices-logo.svg.png" width="800" height="200" />

根据你的训练集和 CPU 速度，你也许能够在本地 CPU 上训练你的神经网络。训练过程可能持续 15 分钟到几个小时，取决于迭代次数、神经网络的大小和其他因素。

一种更快速的方法是在 [GPU (图形处理器)](https://en.wikipedia.org/wiki/Graphics_processing_unit) 上进行训练。虽然 CPU 可以比 GPU 计算更复杂的运算，但是吞吐量较小。因此在训练神经网络时，GPU 比 CPU 的速度要快很多。GPU 在同一时间会执行更多的计算，当我们处理大数据时这很重要。

你也可以自己购买 NVIDIA GPU，但是并非像将 GPU 插入机器那么简单。即使你知道如何安装 GPU，但是正确地安装 CUDNN 还是有一定的挑战。因此，我们不建议这么使用 GPU。相反，我们建议使用 [Amazon EC2 实例](https://aws.amazon.com/cn/ec2/)。

在接下来的几个部分，我们将学习如何从头开始操作，最终在 Amazon 服务器上运行神经网络。

# 4.查看你的上限

Amazon 网络服务具有一项 [弹性计算云 (EC2)](https://en.wikipedia.org/wiki/Amazon_Elastic_Compute_Cloud) 服务，使你能够启动虚拟服务器。这些被 Amazon 称之为“实例”。你要使用的具体实例类型是“g2.2xlarge”。它是一种 GPU 实例，将缩短你在 CPU 上训练神经网络用到的时间。

现在登录你的帐号，并确保你可以登录 GPU 实例。

<img src="../../../sources/img/CNN/ec2-service-limits.png" width="800" height="400" />

- 查看你的 [EC2 服务上限报告](https://us-east-2.console.aws.amazon.com/ec2/v2/home?region=us-east-2#Limits:)
- 找到 g2.2xlarge 实例类型的“当前上限”。

注意：并非每个 AWS 地区都支持 GPU 实例。如果你所选的地区不支持 GPU 实例，但是你想使用 GPU 实例，那么更改你的 AWS 地区。

你应该看到如上所示的 g2.2xlarge 实例上限。在此示例中，上限为 0。默认情况下，AWS 将用户可以运行的 g2.2xlarge 实例数量上限设为 0，意味着你无法启动此实例。如果你看到上限为 0，那么你需要转到下一部分“提交上限提升请求”。否则，你可以跳到下一页的“启动实例”。

### 提交上限提升请求

因为你的 g2.2xlarge 实例上限为 0，所有你需要提高这一上限。在此部分，你将提交请求来提高你的上限。

- 在 EC2 服务上限页面上点击“g2.2xlarge”旁边的“请求升高上限”
  - 你在请求升高上限时不需要缴纳费用。
  
- 填写所有字段。
  - 对于“地区”字段，选择离你最近的地区。
  - 对于 “类型”字段，选择g 或 p 开头的机器（下面列表中都可以，根据地区不同会出现不同的机器以供选择）数字越小越容易通过。
    - g2.xlarge
    - g2.2xlarge
    - g2.4xlarge
    - g2.8xlarge
    - g2.16xlarge
    - g3.xlarge
    - g3.2xlarge
    - g3.4xlarge
    - g3.8xlarge
    - g3.16xlarge
    - p2.xlarge
    - p2.2xlarge
    - p2.4xlarge
    - p2.8xlarge
    - p2.16xlarge
    - p3.xlarge
    - p3.2xlarge
    - p3.4xlarge
    - p3.8xlarge
    - p3.16xlarge
  - 对于“新的上限值”字段，输入 1。
  - 对于“使用情况说明”，输入“Udacity—Nanodegree program”。

### 注意：
如果你从未在 AWS 上启动任何类型的实例，AWS 支持团队可能会向你发送一封电子邮件，要求你通过创建一个实例来初始化你的账号，然后他们才会批准你的上限提升请求。
  
### 等待批准

你必须等待 AWS 批准你的上限提升请求。AWS 通常会在 48 小时内批准这些请求。在等待期间，你可以转到下个部分：申请 AWS Credit。

# 6.启动实例

AWS 批准你的 GPU 上限提升请求后，你就可以开始启动实例流程。

- 转到 [EC2 管理控制台](https://us-east-2.console.aws.amazon.com/ec2/v2/home?region=us-east-2#Home:)

<img src="../../../sources/img/CNN/instance-button.png" width="200" height="600" />

- 点击“启动实例”按钮。
- 点击“Community AMI”
- 搜索“udacity-dl”AMI。
- 针对 udacity-dl AMI 点击“选择”按钮。

[AMI (亚马逊机器镜像)](https://docs.aws.amazon.com/AWSEC2/latest/UserGuide/AMIs.html) 包含在 GPU 上训练神经网络所需的所有环境文件和驱动程序，其中具有 [cuDNN](https://developer.nvidia.com/cudnn)、[TensorFlow](https://www.tensorflow.org/) （支持 GPU）、Python 3，以及本课程所需的所有其他程序包。

### 实例类型

<img src="../../../sources/img/CNN/gpu-instances.png" width="400" height="600" />

- 过滤实例列表，仅显示“GPU 实例”
- 选择 g2.2xlarge 实例类型
- 点击“查看和启动”按钮

### 存储空间大小

你在这门课程中要使用的某些数据集大小可能会达到几个 G。为了处理所有这些数据，你需要有空间来存储这些数据集。确保你的实例将具有 32G 的存储空间。

- 点击“修改存储空间”
- 将存储空间大小增大到 32 GB
- 再次点击“查看和启动”按钮。

### 安全组

在这门课程里，你将运行几个 Jupyter 记事本，默认端口为“8888”。要访问此端口，你需要通过修改安全组在 AWS 上打开此端口。

- 点击“修改安全组”

<img src="../../../sources/img/CNN/aws-add-sec-group.png" width="400" height="600" />

- 设置安全组
  - 选择“创建新的安全组”
  - 将“安全组名称”设为“Jupyter”
  - 点击“添加规则”
  - 设置“自定义 TCP 规则”
  - 将“端口范围”设为“8888”
  - 将“来源”设为“任何地方”
- 点击“查看和启动”

### 启动

- 点击“启动”按钮即可启动 GPU 实例。
  - AWS 将询问你是否要指定身份验证密钥对。此处，AMI 具有预先配置好的用户帐号和密码，所以你可以选择“在没有密钥对的情况下继续”并点击“启动实例”按钮。
- 点击“查看实例”按钮，转到 EC2 管理控制台并观察你的实例开始启动。

### 注意！

从此刻起，如果你运行此 EC2 实例，AWS 会向你收费。详情请参阅 [EC2 按需定价页面](https://amazonaws-china.com/cn/ec2/pricing/on-demand/)。

最重要的是，当你不使用实例时，记得“停止”（即关闭）这些实例。否则，你的实例可能会在你不记得的情况下运行一天、一周、一个月，甚至更长的时间，这样就会产生巨额账单！

AWS 主要对运行实例收费，所以当你停止实例时，大部分的费用就不会产生了。但是，AWS 还是会收取小额的存储空间费用，并且不断积累，直到你“终止”（即删除）实例。

<img src="../../../sources/img/CNN/stop.png" width="400" height="600" />

我们无法设置一定数额的 AWS 预算上限，以便当 AWS 账单达到该阈值时就会自动关闭。但是，你可以 [设置 AWS 账单提醒](http://docs.aws.amazon.com/awsaccountbilling/latest/aboutv2/free-tier-alarms.html)。


# 7.登录

实例可能需要几分钟的时间才能初始化。在 EC2 管理控制台上看到“2/2 检查已通过”消息后，你的实例就可以登录了。

<img src="../../../sources/img/CNN/aws-inst-stats.png" width="400" height="600" />

<center>Instance Status Check and Public IP</center>

注意 EC2 信息中心上的“IPv4 Public IP”地址（格式为“X.X.X.X”。

- 转到终端
- 输入 ssh udacity@X.X.X.X
  - X.X.X.X 是在 AWS 中找到的 IPv4 Public IP。
- 用密码“udacity”进行身份验证。
- 使用 passwd 命令更改密码。
  - 你将使用该新密码登录“udacity”实例。

现在你已登录你的实例。你可以在此实例中运行任何项目了。为了确保一切正常，你可以通过运行示例项目进行测试。
